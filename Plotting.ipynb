```python
def readFileInts(filepath):
    data = []
    with open(filepath) as fp:  
        line = fp.readline().rstrip('\n')
        data.append(int(line))
        while line:
            line = fp.readline().rstrip('\n')
            if line != '':
                data.append(int(line))
    return data
```

```python
def computeMovingAvg(n, data):
    cumsum, moving_aves = [0], []
    for i, x in enumerate(data, 1):
        cumsum.append(cumsum[i-1] + x)
        if i>=N:
            moving_ave = (cumsum[i] - cumsum[i-N])/N
            #can do stuff with moving_ave here
            moving_aves.append(moving_ave)
    return moving_aves


N = 10


def computeAverage(n,data):
    cumsum = []
    nSum = 0
    nIter = 1
    for i, x in enumerate(data, 1):
        nSum += x
        if(i == (nIter*n)):
            nSum = nSum/n
            cumsum.append(nSum)
            nSum = 0
            nIter += 1
        elif (i == len(data)):
            nSum = nSum/n
            cumsum.append(nSum)
            nSum = 0
    return cumsum
        
```
